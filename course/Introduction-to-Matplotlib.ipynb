## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
import matplotlib.pyplot as plt

# 1.prepare date
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. save & show (you save the whole figure)
fig.savefig("../images/Simple-plot.png")
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                            figsize=(10, 5))
# plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
# plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

### which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
# ax.set_xlim([45, 100])
```

```python
## OO method from scratch
fix, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1, 
                               figsize=(10, 10),
                               sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize the plot
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# ADD data to ax1
scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# Customize the plot
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel="Age",
       ylabel="Max Heart Rate")

ax1.legend(*scatter1.legend_elements(), title="Target")

# ADD a horizontal line
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

# ADD a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# customizing the y and x axis limitations

# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1, 
                               figsize=(10, 10),
                               sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter",
                      s=90)
# Set y limits for ax0
ax0.set_ylim([100, 590])

# Customize the plot
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--", color="red")

# ADD data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter",
                       s=90)
# Set x and y limits for ax1
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Customize the plot
ax1.set(title="Heart Disease and Thalach Levels",
       xlabel="Age",
       ylabel="Max Heart Rate")

ax1.legend(*scatter1.legend_elements(), title="Target")

# ADD a horizontal line
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--", color="red")

# ADD a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
